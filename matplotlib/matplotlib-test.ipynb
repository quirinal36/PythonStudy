```python
import pandas as pd
from pandas_datareader import data

# 종목코드, 사이트명, 시작날짜, 종료날짜
# 애플, 야후파이낸스
#df = data.DataReader('AAPL', 'yahoo', start='2021-01-01', end='2021-03-15')

#시각화
#df['Close'].plot()

# 삼성, 네이버파이낸스
df2 = data.DataReader('005930', 'naver', start='2020-01-01', end='2021-01-01')
df2['Close'] = df2['Close'].astype(float)
```

```python
# 차트 사이즈
df2['Close'].plot()
df2['rolling5'] = df2['Close'].rolling(5).mean()
df2['rolling5'].plot()
```

```python
import matplotlib.pyplot as plt

# x축, y축
#plt.plot([1,2,3], [4,5,6])
#plt.show()
plt.figure(figsize=(15,10))
plt.plot(df2.index, df2['Close'], color='skyblue')
plt.xlabel('date')
plt.ylabel('close')
plt.legend(['Samsung'])
```

```python
# 2. 삼성전자와 경쟁업체 LG전자의 매출 데이터는 다음과 같습니다.
date = [2018, 2019, 2020, 2021]
samsung = [50000, 60000, 75000, 70000]
lg = [30000, 40000, 50000, 35000]
my_data = {
    'date':date,
    'samsung':samsung,
    'lg':lg
}
df = pd.DataFrame(my_data)
print(df)

plt.plot(df['date'], df['samsung'])
plt.plot(df['date'], df['lg'])
plt.xlabel('date')
plt.ylabel('value')
plt.legend(['samsung','lg'])
plt.show()
```

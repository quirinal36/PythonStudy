```python
import pandas as pd
from pandas_datareader import data
import FinanceDataReader as fdr
import matplotlib.pyplot as plt

startDt = '2020-01-01'
endDt = '2021-01-01'
currencyDt = fdr.DataReader('USD/KRW',startDt, endDt)

df = data.DataReader('000270', 'naver', start=startDt, end=endDt)
df['Close'] = df['Close'].astype(float) 
#df['Close'].plot()

df2 = data.DataReader('005930', 'naver', start=startDt, end=endDt)
df2['Close'] = df2['Close'].astype(float)
# df2['Close'].plot()

df3 = data.DataReader('003550', 'naver', start=startDt, end=endDt)
df3['Close'] = df3['Close'].astype(float) 

plt.figure(figsize=(12, 8))
plt.plot(df2.index, df2['Close'], color='blue')
plt.plot(df.index, df['Close'], color='red')
plt.plot(df3.index, df3['Close'], color='black')
plt.xlabel('date')
plt.ylabel('close')
plt.legend( ['Samsung', 'Kia', 'LG'] )
plt.show()
```
